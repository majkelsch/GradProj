# Testování v Pythonu — modul `unittest`

Při vývoji softwaru je důležité nejen psát funkce, které plní zamýšlený úkol, ale také zajistit, že budou **vždy fungovat správně** — i po úpravách kódu. Zde přichází na řadu **automatické testování**.

---

## Proč testovat?

| Důvod | Příklad |
|---|---|
| **Odhalení chyb** | Test odhalí, že funkce po refaktoru vrací špatný výsledek |
| **Prevence regresí** | Po přidání nové funkce se neporušila stávající logika |
| **Dokumentace chování** | Testy ukazují, jak se funkce má chovat |
| **Jistota při úpravách** | Můžeme měnit kód a testy nám řeknou, zda vše stále funguje |
| **Úspora času** | Automatické testy jsou rychlejší než ruční ověřování |

## Přehled konceptů v tomto notebooku

| Koncept | Popis |
|---|---|
| `unittest` | Vestavěný modul pro psaní testů |
| `TestCase` | Třída, ze které dědí všechny testovací třídy |
| `assertEqual(a, b)` | Ověří, že `a == b` |
| `assertTrue(x)` / `assertFalse(x)` | Ověří, že `x` je `True` / `False` |
| `assertRaises(Error)` | Ověří, že se vyvolá výjimka |
| `assertIn(a, b)` | Ověří, že `a` je obsaženo v `b` |

## Ruční vs. automatické testování

| | Ruční testování | Automatické testování |
|---|---|---|
| Rychlost | ❌ Pomalé | ✅ Okamžité |
| Opakovatelnost | ❌ Náchylné k chybám | ✅ Vždy stejné |
| Pokrytí | ❌ Snadno přehlédneme | ✅ Systematické |
| Dokumentace | ❌ Žádná | ✅ Testy slouží jako dokumentace |

# Modul `unittest` — základní struktura

Python má vestavěný modul `unittest` — **nemusíme nic instalovat**.

## Jak napsat test?

1. Importujeme `unittest`
2. Vytvoříme třídu, která **dědí** z `unittest.TestCase`
3. Každá testovací metoda **začíná na `test_`**
4. Uvnitř metody používáme **assertovací metody** (`assertEqual`, `assertTrue`, …)
5. Spustíme testy pomocí `unittest.main()`

## Syntaxe

```python
import unittest

class TestMojeFunkce(unittest.TestCase):
    def test_pripad_1(self):
        self.assertEqual(funkce(vstup), ocekavany_vysledek)

    def test_pripad_2(self):
        self.assertTrue(funkce(vstup))

if __name__ == '__main__':
    unittest.main()
```

> **⚠️ Pozor:** V Jupyter Notebooku `unittest.main()` nefunguje správně, protože volá `sys.exit()`. Proto v notebooku vytváříme testovací sadu ručně (viz příklady níže). **Doporučujeme testovat v `.py` souborech.**

### Příklad — první test

````python
import unittest

# Definice funkce, kterou chceme testovat
def secti(x, y):
    return x + y

# Testovací třída — dědí z unittest.TestCase
class TestSectiFunkce(unittest.TestCase):
    def test_kladna_cisla(self):
        self.assertEqual(secti(2, 3), 5)

    def test_zaporna_cisla(self):
        self.assertEqual(secti(-1, 1), 0)

    def test_nuly(self):
        self.assertEqual(secti(0, 0), 0)

# Spuštění testů v Jupyter Notebooku (náhrada za unittest.main())
suite = unittest.TestLoader().loadTestsFromTestCase(TestSectiFunkce)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

> **Tip:** `verbosity=2` zajistí podrobný výpis — uvidíme název každého testu a jeho výsledek (`ok` / `FAIL`).

> **Tip:** `unittest` automaticky spustí **všechny metody**, které začínají na `test_`. Metody bez tohoto prefixu se ignorují.

# Assertovací metody — přehled

Assertovací metody jsou **srdcem testování**. Každá z nich ověřuje jinou podmínku. Pokud podmínka neplatí, test **selže** a zobrazí se chybová zpráva.

## Nejčastější assertovací metody

| Metoda | Ověřuje | Příklad |
|---|---|---|
| `assertEqual(a, b)` | `a == b` | `assertEqual(secti(2, 3), 5)` |
| `assertNotEqual(a, b)` | `a != b` | `assertNotEqual(secti(2, 3), 6)` |
| `assertTrue(x)` | `x is True` | `assertTrue(je_sude(4))` |
| `assertFalse(x)` | `x is False` | `assertFalse(je_sude(3))` |
| `assertIs(a, b)` | `a is b` (stejný objekt) | `assertIs(x, None)` |
| `assertIsNone(x)` | `x is None` | `assertIsNone(vysledek)` |
| `assertIn(a, b)` | `a in b` | `assertIn("a", "ahoj")` |
| `assertNotIn(a, b)` | `a not in b` | `assertNotIn("x", "ahoj")` |
| `assertRaises(Error)` | Vyvolá se výjimka | `assertRaises(ValueError)` |
| `assertAlmostEqual(a, b)` | `a ≈ b` (pro desetinná čísla) | `assertAlmostEqual(0.1 + 0.2, 0.3)` |

> **Pozn.:** Všechny metody přijímají volitelný parametr `msg` pro vlastní chybovou zprávu: `assertEqual(a, b, msg="Hodnoty se neshodují!")`

### Příklad — ukázka různých assertů

````python
import unittest

def je_sude(n):
    return n % 2 == 0

def deleni(a, b):
    if b == 0:
        raise ValueError("Nelze dělit nulou!")
    return a / b

class TestRuzneAsserty(unittest.TestCase):
    def test_assertEqual(self):
        self.assertEqual(2 + 2, 4)

    def test_assertNotEqual(self):
        self.assertNotEqual(2 + 2, 5)

    def test_assertTrue(self):
        self.assertTrue(je_sude(4))

    def test_assertFalse(self):
        self.assertFalse(je_sude(3))

    def test_assertIn(self):
        self.assertIn("a", "ahoj")

    def test_assertRaises(self):
        with self.assertRaises(ValueError):
            deleni(10, 0)

suite = unittest.TestLoader().loadTestsFromTestCase(TestRuzneAsserty)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

> **Tip:** `assertRaises` se používá s kontextovým manažerem `with`. Kód uvnitř bloku `with` by měl vyvolat uvedenou výjimku — pokud se tak nestane, test selže.

# Co se stane, když test selže?

Pokud assertovací metoda zjistí nesoulad, test **selže** (FAIL). Unittest vypíše:
- **Který test selhal** (název metody)
- **Co se očekávalo** vs. **co se skutečně stalo**
- **Traceback** s přesným řádkem

### Příklad — neúspěšný test

````python
import unittest

def secti(x, y):
    return x + y

class TestSectiFunkce(unittest.TestCase):
    def test_spravny(self):
        self.assertEqual(secti(2, 3), 5)  # ✅ projde

    def test_chybny(self):
        self.assertEqual(secti(2, 3), 6)  # ❌ selže — 2+3 = 5, ne 6!

suite = unittest.TestLoader().loadTestsFromTestCase(TestSectiFunkce)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

> **Tip:** Ve výstupu uvidíme `FAIL` u `test_chybny` a podrobnosti: `AssertionError: 5 != 6`. To nám přesně řekne, co bylo očekáváno a co funkce skutečně vrátila.

# Testování hranic vstupů

Důležité je testovat nejen „normální" vstupy, ale i **krajní případy** (boundary testing):
- Co když je vstup **nula**?
- Co když je vstup **záporný**?
- Co když je vstup **na hranici** povoleného rozsahu?

### Příklad — funkce omezující číslo do rozsahu

````python
import unittest

def omez_do_rozsahu(x, min_hodnota, max_hodnota):
    """Omezí číslo x do intervalu [min_hodnota, max_hodnota]."""
    return max(min_hodnota, min(x, max_hodnota))

class TestOmezDoRozsahu(unittest.TestCase):
    def test_hodnota_uvnitr(self):
        self.assertEqual(omez_do_rozsahu(5, 1, 10), 5)

    def test_hodnota_pod_minimum(self):
        self.assertEqual(omez_do_rozsahu(-1, 1, 10), 1)

    def test_hodnota_nad_maximum(self):
        self.assertEqual(omez_do_rozsahu(15, 1, 10), 10)

    def test_hodnota_na_hranici_min(self):
        self.assertEqual(omez_do_rozsahu(1, 1, 10), 1)

    def test_hodnota_na_hranici_max(self):
        self.assertEqual(omez_do_rozsahu(10, 1, 10), 10)

suite = unittest.TestLoader().loadTestsFromTestCase(TestOmezDoRozsahu)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

# Testování práce s textem

Testování řetězců může zahrnovat kontrolu **velkých a malých písmen**, **nahrazování** nebo **obsahu**.

### Příklad — funkce pro záměnu znaků

````python
import unittest

def zamen_znaky(text):
    """Vymění všechna písmena 'a' za 'o'."""
    return text.replace("a", "o")

class TestZamenZnaky(unittest.TestCase):
    def test_zamena_a_na_o(self):
        self.assertEqual(zamen_znaky("banana"), "bonono")

    def test_zamena_na_zacatku(self):
        self.assertEqual(zamen_znaky("apple"), "opple")

    def test_bez_znaku_a(self):
        self.assertEqual(zamen_znaky("kiwi"), "kiwi")

    def test_prazdny_retezec(self):
        self.assertEqual(zamen_znaky(""), "")

suite = unittest.TestLoader().loadTestsFromTestCase(TestZamenZnaky)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

# Testování složitějších datových struktur

Lze otestovat, zda funkce pracuje správně se **seznamy**, **slovníky** či **n-ticemi**. Assertovací metody fungují i pro porovnávání celých kolekcí.

### Příklad — součet odpovídajících prvků dvou seznamů

````python
import unittest

def soucty_prvku(list1, list2):
    """Sečte odpovídající prvky dvou seznamů stejné délky."""
    if len(list1) != len(list2):
        raise ValueError("Seznamy musí mít stejnou délku!")
    return [a + b for a, b in zip(list1, list2)]

class TestSouctyPrvku(unittest.TestCase):
    def test_spravne_scitani(self):
        self.assertEqual(soucty_prvku([1, 2, 3], [4, 5, 6]), [5, 7, 9])

    def test_prazdne_seznamy(self):
        self.assertEqual(soucty_prvku([], []), [])

    def test_nerovnaka_delka(self):
        with self.assertRaises(ValueError):
            soucty_prvku([1, 2], [1, 2, 3])

suite = unittest.TestLoader().loadTestsFromTestCase(TestSouctyPrvku)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

# Testování stavového chování objektů

U objektově orientovaného kódu testujeme, zda objekty **správně mění svůj stav** po volání metod. To je důležité zejména u tříd, které si uchovávají vnitřní data.

### Příklad — třída BankovniUcet

````python
import unittest

class BankovniUcet:
    def __init__(self, pocatecni_zustatek=0):
        self.zustatek = pocatecni_zustatek

    def vklad(self, castka):
        self.zustatek += castka

    def vyber(self, castka):
        if castka > self.zustatek:
            raise ValueError("Nedostatečné prostředky!")
        self.zustatek -= castka

class TestBankovniUcet(unittest.TestCase):
    def test_vklad(self):
        ucet = BankovniUcet(100)
        ucet.vklad(50)
        self.assertEqual(ucet.zustatek, 150)

    def test_vyber(self):
        ucet = BankovniUcet(100)
        ucet.vyber(30)
        self.assertEqual(ucet.zustatek, 70)

    def test_nedostatecny_zustatek(self):
        ucet = BankovniUcet(50)
        with self.assertRaises(ValueError):
            ucet.vyber(100)

    def test_pocatecni_zustatek(self):
        ucet = BankovniUcet()
        self.assertEqual(ucet.zustatek, 0)

suite = unittest.TestLoader().loadTestsFromTestCase(TestBankovniUcet)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
````

# 📋 Shrnutí

| Koncept | Co dělá | Příkaz |
|---|---|---|
| Import | Načte testovací modul | `import unittest` |
| Testovací třída | Seskupuje testy | `class TestX(unittest.TestCase):` |
| Testovací metoda | Jeden test (prefix `test_`) | `def test_pripad(self):` |
| Porovnání | Ověří rovnost | `self.assertEqual(a, b)` |
| Pravdivost | Ověří `True` / `False` | `self.assertTrue(x)` / `self.assertFalse(x)` |
| Výjimka | Ověří vyvolání chyby | `with self.assertRaises(Error):` |
| Obsaženost | Ověří přítomnost prvku | `self.assertIn(a, b)` |
| Spuštění | Spustí všechny testy | `unittest.main()` |

### Kdy použít kterou metodu?

| Situace | Metoda |
|---|---|
| Porovnávám dva výsledky | `assertEqual` |
| Ověřuji, zda funkce vrací `True`/`False` | `assertTrue` / `assertFalse` |
| Funkce má vyhodit chybu při špatném vstupu | `assertRaises` |
| Hledám prvek v seznamu/textu | `assertIn` |
| Porovnávám desetinná čísla | `assertAlmostEqual` |

> 🎯 **Tip na závěr:** Pište testy **krátké a zaměřené** — každá testovací metoda by měla ověřovat **jednu konkrétní věc**. Název metody by měl jasně říkat, co testuje.
