# 1
Vytvořte třídu `Kniha` s těmito atributy (nastavte v `__init__`): **(1b)**

* `nazev`
* `autor`
* `zanr`
* `pocet_stran` — ověřte, že hodnota je `int` (pokud ne, vyvolejte `TypeError`)

Dále implementujte: **(1b)**
* pro `pocet_stran` vytvořte `@property` s `@pocet_stran.setter`, který kontroluje typ
* metodu `__str__`, která vrátí info o knize v čitelném formátu

```python
class Kniha:
    def __init__(self, nazev: str, autor: str, zanr: str, pocet_stran: int):
        self.nazev = nazev
        self.autor = autor
        self.zanr = zanr
        self.pocet_stran = pocet_stran  # projde přes setter

    @property
    def pocet_stran(self) -> int:
        return self._pocet_stran

    @pocet_stran.setter
    def pocet_stran(self, hodnota):
        if not isinstance(hodnota, int):
            raise TypeError(f"Počet stran musí být int, dostal jsem {type(hodnota).__name__}")
        self._pocet_stran = hodnota

    def __str__(self) -> str:
        return f"{self.nazev}, {self.autor}, {self.zanr}, {self.pocet_stran} stran"
    
    def __repr__(self) -> str:
        self.__str__()
```

# 2
Vytvořte třídu `Knihovna` s těmito atributy (nastavte v `__init__`): **(1b)**
* `nazev`
* `knihy` — seznam knih (výchozí prázdný seznam)
* `kapacita` — maximální počet knih

Implementujte tyto metody: **(1b + 2b)**

* `__str__` — vrátí přehled o knihovně (název, obsazenost, výpis knih)
* `__len__` — vrátí aktuální počet knih
* `__contains__(self, kniha)` — umožní použití operátoru `in` (porovnání podle názvu)
* `pridej_knihu(self, kniha)` a `odeber_knihu(self, kniha)` — přidání/odebrání knihy s kontrolou kapacity a duplicit + zpětná vazba pomocí `print()` **(+2b)**

```python
class Knihovna:
    def __init__(self, nazev: str, kapacita: int):
        self.nazev = nazev
        self.kapacita = kapacita
        self.knihy = []

    def __len__(self) -> int:
        return len(self.knihy)

    def __contains__(self, kniha: Kniha) -> bool:
        return any(k.nazev == kniha.nazev for k in self.knihy)

    def pridej_knihu(self, kniha: Kniha) -> None:
        if kniha in self:
            print(f"Kniha '{kniha.nazev}' už v knihovně je.")
        elif len(self) >= self.kapacita:
            print(f"Knihovna je plná ({len(self)}/{self.kapacita}), nelze přidat '{kniha.nazev}'.")
        else:
            self.knihy.append(kniha)
            print(f"Kniha '{kniha.nazev}' byla přidána.")

    def odeber_knihu(self, kniha: Kniha) -> None:
        if kniha not in self:
            print(f"Kniha '{kniha.nazev}' se v knihovně nenachází.")
        else:
            self.knihy = [k for k in self.knihy if k.nazev != kniha.nazev]
            print(f"Kniha '{kniha.nazev}' byla odebrána.")

    def knihy_od_autora(self, autor: str) -> list[Kniha]:
        return [k for k in self.knihy if k.autor == autor]

    def __str__(self) -> str:
        hlavicka = f"V knihovně {self.nazev} je {len(self)}/{self.kapacita} knih."
        if self.knihy:
            seznam = "\n".join(f"  {k}" for k in self.knihy)
            return f"{hlavicka} Jedná se o:\n{seznam}"
        return f"{hlavicka} Knihovna je prázdná."
```

```python
# Založení 5 knih (1b)
k1 = Kniha("Kniha Džunglí", "Dobrovský", "Dobrodružný", 365)
k2 = Kniha("Sparta", "Havlíček", "Historický", 253)
k3 = Kniha("Harry Potter", "Rowlingová", "Fantasy", 723)
k4 = Kniha("1984", "Orwell", "Sci-fi", 311)
k5 = Kniha("Malý princ", "Saint-Exupéry", "Pohádka", 96)

knihy = [k1, k2, k3, k4, k5]

# Výpis všech knih (1b)
print("=== Všechny knihy ===")
for kniha in knihy:
    print(kniha)

# Založení knihovny s kapacitou 3 (1b)
knihovna = Knihovna("Městská knihovna", 3)

# Přidání 2 knih (1b)
print("\n=== Přidávání knih ===")
knihovna.pridej_knihu(k1)
knihovna.pridej_knihu(k2)

# Výpis knih v knihovně (1b)
print("\n=== Stav knihovny ===")
print(knihovna)

# Přidání dalších 2 knih — jedna se nevejde (1b)
print("\n=== Přidání dalších 2 knih ===")
knihovna.pridej_knihu(k3)
knihovna.pridej_knihu(k4)  # kapacita = 3, tato se nevejde

# Výpis celé knihovny (1b)
print("\n=== Výpis knihovny ===")
print(knihovna)

# Knihy od jednoho autora — list comprehension (+1b)
print("\n=== Knihy od autora 'Dobrovský' ===")
od_autora = [k for k in knihovna.knihy if k.autor == "Dobrovský"]
for kniha in od_autora:
    print(kniha)

# Odebrání knihy, která v knihovně není (+1b)
print("\n=== Pokus o odebrání knihy ===")
knihovna.odeber_knihu(k5)  # Malý princ nebyl přidán
```
