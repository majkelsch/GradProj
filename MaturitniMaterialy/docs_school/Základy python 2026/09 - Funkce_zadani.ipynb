# 🔧 Funkce

## Co je to funkce?

**Funkce** je pojmenovaný blok kódu, který provede určitou činnost. Zapíšeme ho **jednou** a pak ho můžeme **volat** (spouštět) kolikrát chceme – z různých míst programu.

### Proč používat funkce?

| Výhoda | Vysvětlení |
|:---|:---|
| **Znovupoužitelnost** | Stejný kód nemusíme psát dvakrát |
| **Přehlednost** | Program rozdělíme na menší, srozumitelné části |
| **Údržba** | Opravíme chybu na jednom místě → opraví se všude |
| **Testovatelnost** | Každou funkci můžeme otestovat samostatně |

> 💡 **Pravidlo:** Pokud někde v programu píšete **stejný kód podruhé**, je čas vytvořit funkci.

### Funkce, které už znáte

V Pythonu jste už několik funkcí používali – jen jste je **nevytvářeli**, ale **volali**:

```python
print("Ahoj")          # vestavěná funkce – vypíše text
len([1, 2, 3])         # vestavěná funkce – vrátí délku
input("Zadej: ")        # vestavěná funkce – načte vstup od uživatele
range(5)                # vestavěná funkce – vygeneruje posloupnost
int("42")               # vestavěná funkce – přetypuje na celé číslo
```

Dnes se naučíme **vytvářet vlastní funkce**.

## 📐 Syntaxe – jak vytvořit funkci

```python
def nazev_funkce(parametr1, parametr2):
    """Popis, co funkce dělá (volitelný, ale doporučený)."""
    příkaz_1
    příkaz_2
    return výsledek   # volitelné – vrátí hodnotu
```

```
┌─────────────────────────────────────────────────┐
│  def   nazev_funkce  (  parametr1, parametr2  ) │
│  ───   ────────────  ─  ─────────────────────  │
│  klíč. jméno funkce  závorky   parametry       │
│  slovo                                          │
├─────────────────────────────────────────────────┤
│      příkaz_1        ← tělo funkce (odsazené)  │
│      příkaz_2                                   │
│      return výsledek ← návratová hodnota        │
└─────────────────────────────────────────────────┘
```

### Pojmy

| Pojem | Význam |
|:---|:---|
| **Definice** (`def`) | Vytvoření funkce – zapíšeme kód funkce |
| **Volání** | Spuštění funkce – `nazev_funkce()` |
| **Parametr** | Proměnná v definici funkce – „místo" pro hodnotu |
| **Argument** | Konkrétní hodnota, kterou předáme při volání |
| **Návratová hodnota** | Výsledek, který funkce vrátí zpět (přes `return`) |
| **Tělo funkce** | Odsazený blok kódu uvnitř funkce |

> ⚠️ **Důležité:** Samotná definice funkce (`def ...`) **nespustí** kód uvnitř! Kód se provede až při **zavolání** funkce.

### Jmenná konvence

V Pythonu existují jasná pravidla pro pojmenování funkcí:

| Pravidlo | ✅ Správně | ❌ Špatně |
|:---|:---|:---|
| Malá písmena + podtržítka (**snake_case**) | `spocitej_prumer` | `SpocitejPrumer`, `spocitejPrumer` |
| Název je **sloveso** (funkce něco *dělá*) | `vypocitej_obvod()` | `obvod()` |
| Popisný název (jasné, co funkce dělá) | `nacti_soubor()` | `funkce1()`, `f()` |
| Bez diakritiky (doporučení) | `spocitej_prumer` | `spočítej_průměr` |

```python
# ✅ Dobře pojmenované funkce:
def vypocitej_bmi(vaha, vyska):
def nacti_data_ze_souboru(cesta):
def je_sude(cislo):              # pro otázky: "je_..."

# ❌ Špatně pojmenované funkce:
def x(a, b):                     # nic neříkající
def MojeFunkce():                # PascalCase → to je konvence pro třídy
def spočítejPrůměr():            # camelCase + diakritika
```

> 💡 **Tip:** Názvy s `je_` / `ma_` / `existuje_` naznačují, že funkce vrací `True`/`False`.

### 🏷️ Type hints – nápověda typů u parametrů

V Pythonu můžete u parametrů funkce **naznačit očekávaný typ** pomocí dvojtečky `:`, a u návratové hodnoty pomocí šipky `->`:

```python
def pozdrav(jmeno: str, vek: int) -> str:
    return f"Ahoj {jmeno}, je ti {vek} let"
```

```
def pozdrav(jmeno: str,  vek: int) -> str:
                   ^^^^       ^^^     ^^^
                   │          │       │
                   │          │       └── návratový typ
                   │          └── parametr typu int
                   └── parametr typu str
```

| Zápis | Význam |
|:------|:-------|
| `jmeno: str` | Parametr `jmeno` by měl být řetězec |
| `vek: int` | Parametr `vek` by měl být celé číslo |
| `-> str` | Funkce vrací řetězec |
| `-> None` | Funkce nic nevrací |

> ⚠️ **Důležité:** Type hints jsou jen **nápověda** – Python je **nekontroluje** za běhu! Funkci můžete klidně předat špatný typ a Python si nebude stěžovat. Slouží hlavně pro:
> - 📖 **Čitelnost** – hned vidíte, co funkce očekává
> - 🔧 **Editor** – VS Code díky nim nabídne napovídání (autocomplete)
> - 🐛 **Hledání chyb** – nástroje jako `mypy` dokáží chyby v typech najít

```python
# Type hints NEZABRÁNÍ předání špatného typu:
def secti(a: int, b: int) -> int:
    return a + b

secti("ahoj", "svete")   # ✅ Python to spustí! → "ahojsvete"
                          # (ale editor/mypy by varoval)
```

## ↩️ `return` – návratová hodnota

Často chceme, aby funkce **vypočítala výsledek** a ten nám **vrátila**. K tomu slouží klíčové slovo `return`:

```python
def secti(a, b):
    vysledek = a + b
    return vysledek    # vrátí hodnotu zpět na místo volání
```

Jakmile Python narazí na `return`:
1. **Okamžitě ukončí** funkci (nic za `return` se neprovede!)
2. **Vrátí hodnotu** na místo, odkud byla funkce zavolána

```
                    Volání                    Definice
                    ──────                    ────────
soucet = secti(3, 5)         ──►  def secti(a, b):     a = 3, b = 5
         │                            return a + b       → 8
         │              ◄────────────────┘
         │
soucet = 8
```

````python
# === return – vrací hodnotu ===
def delka_pozdravu(jmeno):
    pozdrav = f"Nazdar {jmeno}"
    return len(pozdrav)
    print("Toto se NIKDY neprovede – je to za return!")

# Návratovou hodnotu uložíme do proměnné:
pocet_znaku = delka_pozdravu("Emil")
print(f"Pozdrav pro Emila má {pocet_znaku} znaků")

# Nebo ji rovnou použijeme:
print(f"Pozdrav pro Frantu má {delka_pozdravu('Franta')} znaků")
````

> ⚠️ **Pozor:** Pokud funkce **nemá** `return`, vrátí automaticky hodnotu `None`:
>
> ```python
> def pozdrav(jmeno):
>     print(f"Ahoj {jmeno}")
>
> vysledek = pozdrav("Adam")   # vypíše "Ahoj Adam"
> print(vysledek)               # None
> ```

### 🔍 Rozdíl mezi `return` a `print`

Tohle je **nejčastější chyba** začátečníků – zaměňovat `return` a `print`:

| | `print()` | `return` |
|:---|:---|:---|
| **Co dělá** | Vypíše text na obrazovku | Vrátí hodnotu volajícímu kódu |
| **Ukončí funkci?** | ❌ Ne | ✅ Ano |
| **Můžu výsledek uložit?** | ❌ Ne (jen se zobrazí) | ✅ Ano (`x = funkce()`) |
| **Kdy použít** | Pro výpis, ladění | Pro výpočty, zpracování dat |

````python
# === print vs return ===
def spatne_secti(a, b):
    print(a + b)       # jen vypíše, nic nevrátí!

def spravne_secti(a, b):
    return a + b        # vrátí výsledek

# Rozdíl v praxi:
x = spatne_secti(3, 5)     # Vypíše 8, ale x = None!
print(f"x = {x}")          # x = None 😱

y = spravne_secti(3, 5)    # Nic nevypíše, ale y = 8
print(f"y = {y}")          # y = 8 ✅
````

## 📦 Více návratových hodnot

Za `return` můžeme zapsat **víc hodnot** oddělených čárkou. Python je vrátí jako **tuple** (n-tici) a my je můžeme rozbalit do více proměnných:

````python
# === Více návratových hodnot ===
def analyzuj_text(text):
    pocet_znaku = len(text)
    pocet_slov = len(text.split())
    prvni_slovo = text.split()[0] if pocet_slov > 0 else ""
    return pocet_znaku, pocet_slov, prvni_slovo

# Rozbalení do proměnných:
znaky, slova, prvni = analyzuj_text("Python je super jazyk")
print(f"Znaků: {znaky}, Slov: {slova}, První slovo: {prvni}")

# Pokud nás nějaká hodnota nezajímá, použijeme _ :
znaky, _, _ = analyzuj_text("Ahoj světe")
print(f"Jen znaků: {znaky}")
````

## 🏷️ Výchozí hodnoty parametrů

Parametrům můžeme nastavit **výchozí hodnotu** (default). Pokud při volání argument neuvedeme, použije se výchozí:

```python
def pozdrav(jmeno, pozdrav_text="Ahoj"):
    print(f"{pozdrav_text} {jmeno}!")
```

````python
# === Výchozí (default) hodnoty parametrů ===
def mocnina(zaklad, exponent=2):
    return zaklad ** exponent

# S výchozí hodnotou (exponent = 2):
print(mocnina(5))         # 5² = 25
print(mocnina(3))         # 3² = 9

# S vlastní hodnotou:
print(mocnina(2, 10))     # 2¹⁰ = 1024
print(mocnina(5, 3))      # 5³ = 125
````

> ⚠️ **Pravidlo:** Parametry s výchozí hodnotou musí být **na konci** seznamu parametrů:
> ```python
> def f(a, b=10, c=20):    # ✅ správně
> def f(a=10, b, c):       # ❌ SyntaxError!
> ```

### Pojmenované argumenty (keyword arguments)

Při volání funkce můžeme argumenty předat **jménem** – pak nezáleží na pořadí:

````python
# === Pojmenované argumenty ===
def profil(jmeno, vek, mesto="Praha"):
    print(f"{jmeno}, {vek} let, {mesto}")

# Poziční argumenty (záleží na pořadí):
profil("Adam", 17)

# Pojmenované argumenty (nezáleží na pořadí):
profil(vek=22, jmeno="Bára", mesto="Brno")

# Mix (poziční musí být první):
profil("Cyril", vek=19, mesto="Ostrava")
````

### 📦 `*args` a `**kwargs` – proměnný počet argumentů

Někdy nevíme předem, **kolik argumentů** funkce dostane. Python nabízí dvě speciální syntaxe:

| Zápis | Název | Co dělá |
|:------|:------|:--------|
| `*args` | Positional variable args | Sebere **všechny poziční** argumenty navíc do **tuple** |
| `**kwargs` | Keyword variable args | Sebere **všechny pojmenované** argumenty navíc do **slovníku** |

> 💡 Názvy `args` a `kwargs` jsou jen **konvence** – důležité jsou hvězdičky `*` a `**`. Můžete napsat i `*cisla` nebo `**nastaveni`.

#### `*args` – libovolný počet pozičních argumentů

```python
def soucet(*cisla):
    """Sečte libovolný počet čísel."""
    return sum(cisla)      # cisla je tuple, např. (1, 2, 3)

soucet(1, 2)           # → 3
soucet(1, 2, 3, 4, 5)  # → 15
soucet()               # → 0
```

#### `**kwargs` – libovolný počet pojmenovaných argumentů

```python
def vypis_info(**info):
    """Vypíše libovolné informace."""
    for klic, hodnota in info.items():
        print(f"{klic}: {hodnota}")

vypis_info(jmeno="Adam", vek=17, mesto="Praha")
# jmeno: Adam
# vek: 17
# mesto: Praha
```

#### Kombinace – pořadí parametrů

Pokud chcete kombinovat běžné parametry s `*args` a `**kwargs`, musíte dodržet toto pořadí:

```python
def funkce(povinny, volitelny=10, *args, **kwargs):
    ...
```

> ⚠️ **Pořadí:** nejdřív **povinné** → pak **výchozí** → pak `*args` → nakonec `**kwargs`

````python
# === *args – libovolný počet pozičních argumentů ===
def soucet(*cisla):
    """Sečte libovolný počet čísel."""
    print(f"Dostal jsem: {cisla} (typ: {type(cisla).__name__})")
    return sum(cisla)

print(soucet(1, 2))
print(soucet(10, 20, 30, 40))
print(soucet())
print()

# === **kwargs – libovolný počet pojmenovaných argumentů ===
def vypis_info(nadpis, **info):
    """Vypíše informace s nadpisem."""
    print(f"=== {nadpis} ===")
    for klic, hodnota in info.items():
        print(f"  {klic}: {hodnota}")

vypis_info("Student", jmeno="Adam", vek=17, trida="IT2")
print()
vypis_info("Učitel", jmeno="Pan Novák", predmet="Python")
````

## 🔒 Scope – rozsah platnosti proměnných

Funkce je samostatná „krabička". Proměnné vytvořené **uvnitř** funkce **neexistují** mimo ni. Tomuto konceptu říkáme **scope** (rozsah platnosti):

```
┌─────────────── Globální scope ──────────────────┐
│                                                  │
│  jmeno = "Adam"     ← existuje všude             │
│  vek = 20                                        │
│                                                  │
│  ┌──────────── Lokální scope funkce ──────────┐  │
│  │                                            │  │
│  │  def pozdrav():                            │  │
│  │      zprava = f"Ahoj {jmeno}"  ← čte jmeno│  │
│  │      ^^^^^ lokální proměnná               │  │
│  │                                            │  │
│  └────────────────────────────────────────────┘  │
│                                                  │
│  print(zprava)  ← ❌ NameError!                  │
│                    zprava neexistuje venku        │
└──────────────────────────────────────────────────┘
```

### Pravidla scope

| Pravidlo | Příklad |
|:---|:---|
| Funkce **čte** globální proměnné | `PI = 3.14` → ve funkci `print(PI)` ✅ |
| Funkce **nemůže měnit** globální proměnné (bez `global`) | `PI = 3` → ve funkci ❌ |
| Proměnné vytvořené ve funkci **neexistují** venku | `obvod = 2*PI*r` → venku ❌ |
| Klíčové slovo `global` umožní změnu | `global pocet` → pak lze měnit ✅ |

> ⚠️ **`global` se krajně nedoporučuje!** Když se proměnné „náhodně" mění podle zavolané funkce, kód se stává nepřehledným. Raději předávejte hodnoty přes **parametry** a **return**.

````python
# === Scope – ukázka ===
PI = 3.14159  # globální konstanta

def obvod_kruhu(polomer):
    # PI můžeme číst (globální proměnná):
    vysledek = 2 * PI * polomer
    return vysledek     # výsledek vrátíme přes return

o1 = obvod_kruhu(5)
o2 = obvod_kruhu(10)

print(f"Obvod kruhu r=5:  {o1:.2f}")
print(f"Obvod kruhu r=10: {o2:.2f}")

# print(vysledek)  # ❌ NameError – existuje jen uvnitř funkce!
````

````python
# === global – ukázka (NEDOPORUČENO!) ===
pocitadlo = 0

def zvys_pocitadlo():
    global pocitadlo  # „vtáhne" globální proměnnou
    pocitadlo += 1

zvys_pocitadlo()
zvys_pocitadlo()
zvys_pocitadlo()
print(f"Počítadlo: {pocitadlo}")  # 3

# ⚠️ Lepší řešení – bez global:
def zvys(hodnota):
    return hodnota + 1

pocitadlo2 = 0
pocitadlo2 = zvys(pocitadlo2)
pocitadlo2 = zvys(pocitadlo2)
pocitadlo2 = zvys(pocitadlo2)
print(f"Počítadlo2: {pocitadlo2}")  # 3 – stejný výsledek, čistší kód
````

## 🧠 Dobré praktiky

### 1. Jedno pravidlo, jedna funkce

Každá funkce by měla dělat **jednu věc** a dělat ji dobře:

```python
# ❌ Špatně – funkce dělá příliš mnoho:
def zpracuj_studenta(jmeno, znamky):
    prumer = sum(znamky) / len(znamky)
    if prumer < 2:
        print(f"{jmeno} je výborný student")
    uloz_do_databaze(jmeno, prumer)

# ✅ Lépe – rozdělíme na menší funkce:
def spocitej_prumer(znamky):
    return sum(znamky) / len(znamky)

def ohodnot_studenta(jmeno, prumer):
    if prumer < 2:
        print(f"{jmeno} je výborný student")
```

Hned pod hlavičku funkce můžete přidat **trojitý string** s popisem. Ten se pak zobrazí v nápovědě:

def spocitej_bmi(vaha_kg, vyska_m):

## ❌ Typické chyby začátečníků

### 1. Zapomenuté závorky při volání

```python
def pozdrav():
    print("Ahoj!")

pozdrav    # ❌ Nic se nestane! Pouze odkaz na funkci
pozdrav()  # ✅ Zavolá funkci → vypíše "Ahoj!"
```

### 2. Volání funkce PŘED její definicí

```python
pozdrav()  # ❌ NameError – Python ještě funkci nezná!

def pozdrav():
    print("Ahoj!")
```

> ⚠️ Python čte kód **odshora dolů**. Funkce musí být **definovaná dříve**, než ji zavoláte.

### 3. Přepsání vestavěné funkce

```python
print = 5          # ❌ Přepíšete vestavěnou funkci print!
print("Ahoj")      # TypeError: 'int' object is not callable

# Řešení: del print   (obnoví vestavěnou funkci)
```

| Chyba | Příčina | Řešení |
|:---|:---|:---|
| `NameError` | Funkce ještě není definovaná | Přesuňte `def` **nad** volání |
| Nic se nestane | Chybí `()` při volání | `pozdrav()` místo `pozdrav` |
| `TypeError: not callable` | Přepsali jste vestavěnou funkci | Nepoužívejte `print`, `len`, `input` jako názvy proměnných |

## 📋 Shrnutí

### Syntaxe funkce

```python
def nazev(parametr1, parametr2="výchozí"):
    """Popis funkce."""
    # tělo funkce
    return vysledek
```

### Přehled pojmů

| Pojem | Význam | Příklad |
|:---|:---|:---|
| `def` | Definice (vytvoření) funkce | `def secti(a, b):` |
| Parametr | Proměnná v definici | `a`, `b` v `def secti(a, b)` |
| Argument | Hodnota při volání | `3`, `5` v `secti(3, 5)` |
| `return` | Vrátí hodnotu + ukončí funkci | `return a + b` |
| Výchozí hodnota | Default parametr | `def f(x=10):` |
| Keyword argument | Pojmenovaný argument | `f(x=5, y=10)` |
| Scope | Rozsah platnosti proměnné | Lokální vs. globální |
| Docstring | Dokumentační řetězec | `"""Popis funkce."""` |

### `return` vs `print`

| | `return` | `print()` |
|:---|:---|:---|
| Vrátí hodnotu | ✅ Ano | ❌ Ne (vrátí `None`) |
| Ukončí funkci | ✅ Ano | ❌ Ne |
| Výsledek lze uložit | ✅ `x = f()` | ❌ Jen zobrazí |

### Scope – pravidla

| Pravidlo | |
|:---|:---|
| Lokální proměnná neexistuje vně funkce | `x` definovaná ve funkci → venku ❌ |
| Funkce čte globální proměnné | `PI = 3.14` → ve funkci ✅ |
| Pro změnu globální proměnné musí být `global` | `global x` → pak lze měnit (nedoporučeno!) |
| Lepší je předávat přes parametry a `return` | Čistší, přehlednější, bezpečnější |

> 🎯 **Zapamatuj si:** Funkce = pojmenovaný blok kódu, který přijímá **parametry**, provede výpočet a vrátí **výsledek** přes `return`. Rozdělení kódu do funkcí je základ dobrého programování!
