## Proč Flask?

|---|---|
| **Jednoduchost** | Funkční web na 5 řádků kódu |
| **Flexibilita** | Nepřikazuje strukturu projektu |

# Základy Flask — vytvoření aplikace

Každá Flask aplikace začíná vytvořením instance třídy `Flask` a definicí **route** (cest).

## Syntaxe

```python
from flask import Flask

app = Flask(__name__)  # vytvoření aplikace

@app.route('/')        # dekorátor — definuje URL cestu
def home():            # funkce, která se zavolá při návštěvě
    return "Ahoj, světe!"

if __name__ == '__main__':
    app.run()           # spustí vývojový server (http://127.0.0.1:5000)
```

### Klíčové pojmy

| Pojem | Popis |
|---|---|
| `Flask(__name__)` | Vytvoří instanci aplikace |
| `@app.route('/')` | **Dekorátor** — mapuje URL na funkci |
| `return` | Vrátí odpověď prohlížeči (text, HTML…) |
| `app.run()` | Spustí vývojový server na `http://127.0.0.1:5000` |

> **Tip:** `__name__` říká Flasku, kde hledat šablony a statické soubory. Vždy ho použijte.

# 1️⃣ Route `/1` — vrácení textu

Nejjednodušší route — funkce vrátí **prostý text** přímo z Pythonu, bez jakéhokoli HTML souboru.

### Kód

````python
# Route /1 — vrácení textu přímo z Pythonu
from flask import Flask

app = Flask(__name__)

@app.route('/1')
def pozdrav_returnem():
    return "Ahoj, světe od jedničky!"

# Po spuštění: http://127.0.0.1:5000/1
````

> **Pozn.:** Text se zobrazí přímo v prohlížeči bez jakéhokoli formátování. Pro hezčí výstup potřebujeme HTML šablony — viz další sekce.

# 2️⃣ Route `/2` — HTML šablona (`render_template`)

Funkce `render_template()` načte HTML soubor ze složky `templates/` a vrátí ho prohlížeči.

## Syntaxe

```python
from flask import render_template

@app.route('/2')
def stranka():
    return render_template("soubor.html")
```

> **⚠️ Pozor:** HTML soubory **musí** být ve složce `templates/` — Flask je hledá automaticky právě tam.

### Kód

````python
# Route /2 — načtení HTML šablony
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/2')
def pozdrav_ze_souboru():
    return render_template("index2.html")
````

### Obsah `templates/index2.html`

```html
<head>
</head>
<body>
    Ahoj, světe ze souboru!
</body>
```

> **Tip:** Šablona je obyčejný HTML soubor. Flask ho načte, vrátí prohlížeči a ten ho zobrazí.

# 3️⃣ Route `/3` — CSS stylování (statické soubory)

CSS soubory, obrázky a další **statické soubory** se ukládají do složky `static/`. Pro odkaz na ně v HTML používáme funkci `url_for()`.

## Syntaxe v HTML (Jinja2)

```html
<link rel="stylesheet" href="{{ url_for('static', filename='styles/styles.css') }}">
```

| Část | Význam |
|---|---|
| `url_for('static', ...)` | Vygeneruje správnou cestu ke statickému souboru |
| `filename='styles/styles.css'` | Relativní cesta uvnitř složky `static/` |
| `{{ ... }}` | Jinja2 výraz — vyhodnotí se na straně serveru |

### Kód

````python
# Route /3 — HTML se stylováním pomocí CSS
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/3')
def pozdrav_ze_souboru_CSS():
    return render_template("index3.html")
````

### Obsah `templates/index3.html`

```html
<html>
    <head>
        <link rel="stylesheet" href="{{ url_for('static', filename='styles/styles.css') }}">
    </head>
    <body>
        <h1>Ahoj světe ze souboru</h1>
    </body>
</html>
```

### Obsah `static/styles/styles.css`

```css
h1 {
    color: blue;
}
```

> **Tip:** Nikdy nepište cestu ke statickým souborům napevno (`/static/styles.css`). Vždy používejte `url_for('static', filename=...)` — funguje správně i při nasazení na server.

# 4️⃣ Route `/4` — předání proměnné do šablony

Jednou z největších výhod Flasku je možnost **předávat data z Pythonu do HTML**. Data předáme jako pojmenované argumenty funkce `render_template()`.

## Syntaxe

```python
# Python (app.py)
return render_template("stranka.html", nazev_v_html=hodnota_z_pythonu)
```

```html
<!-- HTML (šablona) -->
<h1>{{ nazev_v_html }}</h1>
```

| Část | Kde | Význam |
|---|---|---|
| `message=text` | Python | Předání proměnné `text` pod názvem `message` |
| `{{ message }}` | HTML | Vypsání hodnoty proměnné `message` |

### Kód

````python
# Route /4 — předání proměnné z Pythonu do HTML
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/4')
def pozdrav_z_promenny():
    text = "Ahoj z proměnné"
    return render_template("index4.html", message=text)
````

### Obsah `templates/index4.html`

```html
<html>
    <head>
        <link rel="stylesheet" href="{{ url_for('static', filename='styles/styles.css') }}">
    </head>
    <body>
        <h1>{{ message }}</h1>
    </body>
</html>
```

> **Tip:** Můžete předat libovolný počet proměnných: `render_template("s.html", jmeno=jm, vek=v, seznam=s)`

# 6️⃣ Route `/6` — formuláře (GET & POST)

Formuláře umožňují uživateli **odesílat data** na server. Flask rozlišuje dvě HTTP metody:

| Metoda | Kdy se použije | Jak funguje |
|---|---|---|
| `GET` | Při **otevření** stránky | Data v URL (viditelná) |
| `POST` | Při **odeslání** formuláře | Data v těle požadavku (skrytá) |

## Syntaxe

```python
@app.route('/6', methods=['GET', 'POST'])
def formular():
    if request.method == 'POST':          # uživatel odeslal formulář
        hodnota = request.form.get('nazev_pole')  # přečti hodnotu
    return render_template('formular.html')
```

```html
<form method="post">
    <input type="number" name="nazev_pole">
    <button type="submit">Odeslat</button>
</form>
```

| Část | Význam |
|---|---|
| `methods=['GET', 'POST']` | Route přijímá obě metody |
| `request.method == 'POST'` | Kontrola, zda uživatel odeslal formulář |
| `request.form.get('number')` | Čtení hodnoty z formuláře (atribut `name`) |
| `type=int` | Automatická konverze na celé číslo |

### Kód

````python
# Route /6 — formulář: přičtení jedničky k zadanému číslu
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/6', methods=['GET', 'POST'])
def prvniFormularCislo():
    result = None
    if request.method == 'POST':
        number = request.form.get('number', type=int)  # přečti z formuláře
        if number is not None:
            result = number + 1  # přičtení jedničky
    return render_template('index6.html', result=result)
````

### Obsah `templates/index6.html`

```html
<!DOCTYPE html>
<html lang="cs">
<head>
    <meta charset="UTF-8">
    <title>Přičti Jedničku</title>
    <link rel="stylesheet" href="{{ url_for('static', filename='styles/styles.css') }}">
</head>
<body>
    <h1>Přičti Jedničku</h1>
    <form method="post">
        <label for="number">Zadej číslo:</label>
        <input type="number" name="number" required>
        <button type="submit">Odeslat</button>
    </form>

    {% if result is not none %}
        <h2>Výsledek: {{ result }}</h2>
    {% endif %}
</body>
</html>
```

### Jinja2 — podmínky v šablonách

| Zápis | Význam |
|---|---|
| `{{ proměnná }}` | Vypsání hodnoty proměnné |
| `{% if podmínka %}` | Podmínka (if/elif/else) |
| `{% endif %}` | Ukončení podmínky |
| `{% for x in seznam %}` | Cyklus |
| `{% endfor %}` | Ukončení cyklu |

> **Tip:** Atribut `name="number"` v HTML musí odpovídat klíči v `request.form.get('number')` v Pythonu.

# 7️⃣ Route `/7` — nahrávání souborů

Flask umožňuje nahrávat soubory přes formulář. Klíčový je atribut `enctype="multipart/form-data"` v HTML.

## Syntaxe

```python
# Nastavení složky pro upload
app.config["UPLOAD_FOLDER"] = "static/uploadedFiles/"
os.makedirs(app.config["UPLOAD_FOLDER"], exist_ok=True)

# Čtení souboru
file = request.files.get('file')
file.save(cesta)              # uložení na disk
file.seek(0)                  # reset pozice pro čtení
obsah = file.read().decode('utf-8')  # přečtení obsahu
```

```html
<form method="post" enctype="multipart/form-data">
    <input type="file" name="file" accept=".txt" required>
    <button type="submit">Odeslat</button>
</form>
```

| Část | Význam |
|---|---|
| `enctype="multipart/form-data"` | Povolí odesílání souborů ve formuláři |
| `request.files.get('file')` | Získání nahraného souboru |
| `file.save(cesta)` | Uloží soubor na disk |
| `file.seek(0)` | Vrátí čtecí pozici na začátek (po `save`) |

### Kód

````python
# Route /7 — nahrání a přečtení textového souboru
from flask import Flask, render_template, request
import os

app = Flask(__name__)

# Vytvoření složky pro upload
app.config["UPLOAD_FOLDER"] = "static/uploadedFiles/"
os.makedirs(app.config["UPLOAD_FOLDER"], exist_ok=True)
@app.route('/7', methods=['GET', 'POST'])
def nahrani_souboru():
    content = None
    if request.method == 'POST':
        file = request.files.get('file')
        if file and file.filename.endswith('.txt'):
            # Uložení souboru na disk
            file_path = os.path.join(app.config["UPLOAD_FOLDER"], file.filename)
            file.save(file_path)
            # Čtení obsahu souboru
            file.seek(0)  # reset pozice po uložení
            content = file.read().decode('utf-8')
    return render_template('index7.html', content=content)
````

### Obsah `templates/index7.html`
```html
<!DOCTYPE html>
<html lang="cs">
<body>
    <h1>Nahrát TXT Soubor</h1>
    <form method="post" enctype="multipart/form-data">
        <label for="file">Vyber TXT soubor:</label>
        <input type="file" name="file" accept=".txt" required>
        <button type="submit">Odeslat</button>
    </form>
    {% if content %}
        <h2>Obsah souboru:</h2>
        <pre>{{ content }}</pre>
    {% endif %}
</body>
</html>
```

> **Tip:** Značka `<pre>` zachovává formátování textu (mezery, nové řádky).

# 8️⃣ Route `/8` — interaktivní grafy (Plotly)

Knihovna **Plotly** umožňuje vytvářet interaktivní grafy přímo v prohlížeči. Graf vytvoříme v Pythonu, převedeme ho na HTML a vložíme do šablony.

## Syntaxe

```python
import plotly.graph_objects as go
import plotly.io as pio

fig = go.Figure()
fig.add_trace(go.Scatter(x=[...], y=[...], mode='lines+markers', name='Data'))
fig.update_layout(title="Název", xaxis_title="X", yaxis_title="Y")

graph_html = pio.to_html(fig, full_html=False)  # převod na HTML fragment
return render_template("graf.html", graph_html=graph_html)
```

```html
<!-- V šabloně: safe filtr zajistí, že se HTML nevyescapuje -->
<div>{{ graph_html|safe }}</div>
```

| Typ grafu | Třída | Příklad |
|---|---|---|
| Čárový / bodový | `go.Scatter` | `mode='lines+markers'` |
| Sloupcový | `go.Bar` | `go.Bar(x=[...], y=[...])` |
| Koláčový | `go.Pie` | `go.Pie(labels=[...], values=[...])` |

### Kód — čárový graf

````python
# Route /8 — interaktivní čárový graf (Plotly)
from flask import Flask, render_template
import plotly.graph_objects as go
import plotly.io as pio

app = Flask(__name__)

@app.route('/8')
def graph():
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=[1, 2, 3, 4], y=[10, 20, 25, 30],
                             mode='lines+markers', name='Data 1'))
    fig.add_trace(go.Scatter(x=[1, 2, 3, 4], y=[15, 18, 22, 27],
                             mode='lines+markers', name='Data 2'))

    fig.update_layout(
        title="Ukázkový interaktivní graf",
        xaxis_title="X-osa",
        yaxis_title="Y-osa",
        template="plotly_white"
    )

    graph_html = pio.to_html(fig, full_html=False)  # HTML fragment
    return render_template("index8.html", graph_html=graph_html)
````

### Kód — sloupcový graf (Route `/8_2`)

````python
# Route /8_2 — sloupcový graf s více sériemi
@app.route('/8_2')
def graphBar():
    fruit = ["Apples", "Oranges", "Bananas", "Apples", "Oranges", "Bananas"]
    contestant = ["Alex", "Alex", "Alex", "Jordan", "Jordan", "Jordan"]
    number_eaten = [2, 1, 3, 1, 3, 2]

    unique_contestants = list(set(contestant))

    fig = go.Figure()
    for current_contestant in unique_contestants:
        x = [fruit[i] for i in range(len(fruit)) if contestant[i] == current_contestant]
        y = [number_eaten[i] for i in range(len(number_eaten)) if contestant[i] == current_contestant]
        fig.add_trace(go.Bar(x=x, y=y, name=current_contestant))

    fig.update_layout(
        title="Number of Fruits Eaten by Contestants",
        legend_title_text="Contestant"
    )
    fig.update_xaxes(title_text="Fruit")
    fig.update_yaxes(title_text="Number Eaten")

    graph_html = pio.to_html(fig, full_html=False)
    return render_template("index8.html", graph_html=graph_html)
````

### Obsah `templates/index8.html`

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Graf ve Flasku</title>
</head>
<body>
    <h1>Jednoduchý interaktivní graf ve Flasku</h1>
    <div>{{ graph_html|safe }}</div>
</body>
</html>
```

> **⚠️ Pozor:** Filtr `|safe` je nutný — říká Jinja2, aby HTML kód grafu **nevyescapoval**. Bez něj by se zobrazil jako text.

# 9️⃣ Route `/9` — parametry v URL

Flask umožňuje definovat **dynamické části** URL pomocí `< >`. Hodnoty z URL se předají jako argumenty funkce.

## Syntaxe

```python
@app.route('/cesta/<int:id>/<string:name>')
def funkce(id, name):
    ...
```

### Typy konvertorů

| Konvertor | Popis | Příklad URL |
|---|---|---|
| `<string:name>` | Textový řetězec (výchozí) | `/user/Adam` |
| `<int:id>` | Celé číslo | `/user/42` |
| `<float:price>` | Desetinné číslo | `/product/19.99` |
| `<path:subpath>` | Cesta (včetně `/`) | `/files/docs/readme.txt` |

### Kód

````python
# Route /9 — parametry z URL adresy
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/9/<int:id>/<string:name>', methods=['GET'])
def parametry(id, name):
    return render_template("index9.html", id=id, name=name)

# Příklad: http://127.0.0.1:5000/9/1/Adam
# → id=1, name="Adam"
````

### Obsah `templates/index9.html`

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Parametry z adresy</title>
    <link rel="stylesheet" href="{{ url_for('static', filename='styles/styles.css') }}">
</head>
<body>
    Vypsání uživatele z parametrů adresy <br>
    <h1>ID: {{ id }} a jeho jméno je: {{ name }}</h1>
</body>
</html>
```

> **Tip:** Parametry v URL se často používají pro zobrazení detailu (např. `/produkt/42` zobrazí produkt s ID 42).

### 9.2 — Query string parametry (`?klíč=hodnota`)

Kromě parametrů přímo v cestě (`/9/1/Adam`) existují i **query string** parametry — ty se píší za `?` na konci URL a oddělují se `&`:

```
http://127.0.0.1:5000/hledej?jmeno=Adam&vek=18
```

K jejich čtení slouží `request.args.get()`.

## Syntaxe

```python
from flask import request

@app.route('/hledej')
def hledej():
    jmeno = request.args.get('jmeno')           # → "Adam"
    vek = request.args.get('vek', type=int)      # → 18 (jako int)
    jazyk = request.args.get('jazyk', 'cs')      # → "cs" (výchozí hodnota)
```

### Srovnání: path vs. query parametry

| | Path parametry | Query string parametry |
|---|---|---|
| **URL** | `/user/42/Adam` | `/user?id=42&jmeno=Adam` |
| **Dekorátor** | `@app.route('/user/<int:id>/<name>')` | `@app.route('/user')` |
| **Čtení** | Argumenty funkce `def f(id, name)` | `request.args.get('id')` |
| **Povinné?** | Ano — bez nich route neodpovídá | Ne — můžou chybět |
| **Typický účel** | Identifikace zdroje (detail) | Filtrování, vyhledávání, stránkování |

### Kód

````python
# Route /hledej — query string parametry z URL (?klíč=hodnota)
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route('/hledej')
def hledej():
    jmeno = request.args.get('jmeno', '')          # výchozí prázdný string
    vek = request.args.get('vek', type=int)         # automatická konverze na int
    trida = request.args.get('trida', 'nezadáno')   # výchozí hodnota

    return render_template("hledej.html", jmeno=jmeno, vek=vek, trida=trida)

# Příklady URL:
# /hledej?jmeno=Adam              → jmeno="Adam", vek=None, trida="nezadáno"
# /hledej?jmeno=Eva&vek=18        → jmeno="Eva",  vek=18,   trida="nezadáno"
# /hledej?jmeno=Petr&trida=EP4A   → jmeno="Petr", vek=None, trida="EP4A"
````

### Ukázka `templates/hledej.html`

```html
<!DOCTYPE html>
<html lang="cs">
<head>

    <title>Hledání</title>
</head>
<body>

        <p>Jméno: <strong>{{ jmeno }}</strong></p>
    {% endif %}

    {% if vek %}
        <p>Věk: <strong>{{ vek }}</strong></p>
    {% endif %}

    <p>Třída: <strong>{{ trida }}</strong></p>

    <hr>
    <h3>Zkus zadat parametry do URL:</h3>
    <ul>
        <li><a href="/hledej?jmeno=Adam">/hledej?jmeno=Adam</a></li>
```

### Další příklad — stránkování výsledků

Query parametry se běžně používají pro **stránkování** (`?strana=2`) nebo **řazení** (`?sort=jmeno&order=asc`):

````python
# Praktická ukázka — stránkování seznamu
@app.route('/produkty')
def produkty():
    vsechny_produkty = ["Notebook", "Myš", "Klávesnice", "Monitor",
                        "Sluchátka", "Webkamera", "USB hub", "SSD disk"]

    strana = request.args.get('strana', 1, type=int)  # výchozí = 1
    na_stranku = 3  # kolik položek na stránku

    # Výpočet rozsahu
    start = (strana - 1) * na_stranku
    end = start + na_stranku
    produkty_na_strance = vsechny_produkty[start:end]

    celkem_stran = (len(vsechny_produkty) + na_stranku - 1) // na_stranku

    return render_template("produkty.html",
                           produkty=produkty_na_strance,
                           strana=strana,
                           celkem_stran=celkem_stran)

# /produkty          → strana 1: Notebook, Myš, Klávesnice
# /produkty?strana=2 → strana 2: Monitor, Sluchátka, Webkamera
# /produkty?strana=3 → strana 3: USB hub, SSD disk
````

> **Tip:** `request.args` je slovníkový objekt obsahující **všechny** query parametry. Metoda `.get()` je bezpečná — pokud parametr v URL chybí, vrátí `None` (nebo zadanou výchozí hodnotu). Na rozdíl od `request.args['klíč']`, která by vyhodila chybu `400 Bad Request`.

# 🔟 Route `/10` — přesměrování (`redirect`)

Funkce `redirect()` přesměruje uživatele na **jinou URL adresu**. Často se kombinuje s `url_for()`, která vygeneruje URL podle názvu funkce.

## Syntaxe

```python
from flask import redirect, url_for

return redirect('/1')              # přesměrování na konkrétní URL
return redirect(url_for('home'))   # přesměrování podle názvu funkce
```

| Funkce | Popis |
|---|---|
| `redirect('/cesta')` | Přesměruje na zadanou URL |
| `url_for('nazev_funkce')` | Vygeneruje URL podle názvu Python funkce |

> **Tip:** Preferujte `url_for()` před pevnými cestami — pokud změníte URL v dekorátoru, `url_for()` se automaticky přizpůsobí.

### Kód

````python
# Route /10 — přesměrování na základě vstupu z formuláře
from flask import Flask, render_template, request, redirect

app = Flask(__name__)

@app.route('/10', methods=['GET', 'POST'])
def redirekting():
    result = None
    if request.method != 'POST':
        return render_template("index10.html", result=result)

    number = request.form.get('number', type=int)
    result = number

    if result == 1:
        return redirect('/1')   # přesměruj na route /1
    elif result == 2:
        return redirect('/2')   # přesměruj na route /2
    else:
        return render_template("index10.html", result=result)
````

> **Pozn.:** Pokud uživatel zadá 1 nebo 2, je přesměrován na odpovídající stránku. Jinak zůstane na stránce `/10` a zobrazí se zadané číslo.

# 1️⃣1️⃣ Registrace a Login — autentizace uživatelů

Pro registraci a přihlašování uživatelů potřebujeme:

| Komponenta | Popis |
|---|---|
| **MySQL** | Databáze pro uložení uživatelů |
| **bcrypt** | Hashování hesel (nikdy neukládáme hesla v čitelné podobě!) |
| **session** | Udržení přihlášeného stavu mezi požadavky |
| **secret_key** | Tajný klíč pro šifrování session dat |

## Připojení k databázi

Přihlašovací údaje importujeme z modulu `pripojeni.py`:
# pripojeni.py

```python
from pripojeni import *
import mysql.connector

mydb = mysql.connector.connect(
    host=HOST, user=USER, password=PASSWORD, database=DATABASE
)
```

## Hashování hesel (bcrypt)

```python
import bcrypt

# Registrace — hashování hesla
hashed = bcrypt.hashpw(heslo.encode('utf-8'), bcrypt.gensalt())
do_db = hashed.decode('utf-8')  # uložíme jako string

# Login — ověření hesla
bcrypt.checkpw(zadane_heslo.encode('utf-8'), hash_z_db.encode('utf-8'))  # → True/False
```

| Funkce | Popis |
|---|---|
| `bcrypt.hashpw()` | Vytvoří hash hesla (jednosměrný) |
| `bcrypt.gensalt()` | Vygeneruje náhodnou "sůl" (ochrana proti rainbow tables) |
| `bcrypt.checkpw()` | Porovná zadané heslo s hashem |

> **⚠️ Pozor:** Hesla se **nikdy** neukládají v čitelné podobě! Vždy se hashují.

## Session — udržení přihlášení

```python
app.secret_key = 'Muj_tajny_klic'  # nutné pro šifrování session

session['email'] = email           # uložení do session (přihlášení)
session.pop('email', None)         # odebrání ze session (odhlášení)
session.get('email')               # čtení ze session
```

### Kód — registrace

````python
# Route /register — registrace nového uživatele
from flask import Flask, render_template, request, redirect, url_for, session
import bcrypt
import mysql.connector
from pripojeni import *

app = Flask(__name__)
app.secret_key = 'Muj_tajny_klic'

@app.route('/register', methods=['GET', 'POST'])
def register():
    if request.method == 'POST':
        name = request.form['jmeno']
        mail = request.form['email']
        psw = request.form['psw']

        # Hashování hesla
        hashed_password = bcrypt.hashpw(psw.encode('utf-8'), bcrypt.gensalt())
        hesloDoDB = hashed_password.decode('utf-8')

        # Připojení k DB
        mydb = mysql.connector.connect(
            host=HOST, user=USER, password=PASSWORD, database=DATABASE
        )
        mycursor = mydb.cursor()

        # Vytvoření tabulky (pokud neexistuje)
        mycursor.execute("""CREATE TABLE IF NOT EXISTS uzivatele (
            id int AUTO_INCREMENT PRIMARY KEY,
            jmeno varchar(35) NOT NULL,
            email varchar(50) NOT NULL,
            heslo varchar(255) NOT NULL
        );""")
        mydb.commit()

        # Vložení uživatele
        sql = "INSERT INTO uzivatele (jmeno, email, heslo) VALUES (%s, %s, %s)"
        mycursor.execute(sql, (name, mail, hesloDoDB))
        mydb.commit()

        return redirect(url_for('login'))  # po registraci → přihlášení

    return render_template("register.html")
````

### Kód — login

````python
# Route /login — přihlášení uživatele
@app.route('/login', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        email = request.form['email']
        password = request.form['psw']

        mydb = mysql.connector.connect(
            host=HOST, user=USER, password=PASSWORD, database=DATABASE
        )
        mycursor = mydb.cursor()
        mycursor.execute("SELECT heslo FROM uzivatele WHERE email = %s;", (email,))
        result = mycursor.fetchone()

        if result:
            stored_hash = result[0]
            if bcrypt.checkpw(password.encode('utf-8'), stored_hash.encode('utf-8')):
                session['email'] = email  # uložení do session
                return redirect(url_for('home_login_ukazka'))
            else:
                error_message = "Nesprávné heslo."
        else:
            error_message = "Uživatel nenalezen."

        return render_template("login.html", error=error_message)

    return render_template("login.html")
````

### Kód — odhlášení a domovská stránka

````python
# Route /home — domovská stránka (zobrazí email přihlášeného uživatele)
@app.route('/home')
def home_login_ukazka():
    return render_template('home.html', email=session.get('email'))

# Route /logout — odhlášení uživatele
@app.route('/logout')
def logout():
    session.pop('email', None)  # odebrání ze session
    return redirect(url_for('home_login_ukazka'))
````

> **Tip:** V šabloně `home.html` pomocí `{% if not email %}` rozlišujeme, zda je uživatel přihlášen — pokud ano, zobrazíme jeho email a tlačítko Logout; pokud ne, zobrazíme odkazy na Login a Register.

# 1️⃣3️⃣ Chybové stránky (`errorhandler`, `abort`)

Když uživatel navštíví neexistující URL, Flask zobrazí výchozí chybovou stránku. Pomocí `@app.errorhandler()` můžeme vytvořit **vlastní** chybové stránky. Funkcí `abort()` můžeme chybu vyvolat sami.

## Syntaxe

```python
from flask import abort

@app.errorhandler(404)
def stranka_nenalezena(error):
    return render_template("404.html"), 404  # pozor — vracíme i status kód!

@app.route('/admin')
def admin():
    abort(403)  # ručně vyvolá chybu 403 Forbidden
```

### Nejčastější HTTP chybové kódy

| Kód | Název | Kdy se použije |
|:---:|---|---|
| `400` | Bad Request | Chybný požadavek (špatná data z formuláře) |
| `403` | Forbidden | Přístup zakázán (nemáte oprávnění) |
| `404` | Not Found | Stránka neexistuje |
| `405` | Method Not Allowed | Špatná HTTP metoda (GET místo POST) |
| `500` | Internal Server Error | Chyba na serveru (bug v kódu) |

> **⚠️ Pozor:** Za `return render_template(...)` musíte přidat **čárku a status kód** (např. `, 404`). Jinak prohlížeč dostane odpověď 200 OK, i když zobrazujete chybovou stránku.

### Kód

````python
# Vlastní chybové stránky — errorhandler
from flask import Flask, render_template, abort

app = Flask(__name__)

# Vlastní stránka pro chybu 404
@app.errorhandler(404)
def page_not_found(error):
    return render_template("404.html"), 404  # vrátí HTML + status kód 404

# Vlastní stránka pro chybu 500
@app.errorhandler(500)
def internal_error(error):
    return render_template("500.html"), 500

# Příklad — ručně vyvolat chybu
@app.route('/tajne')
def tajna_stranka():
    abort(403)  # → vyvolá chybu 403 Forbidden
````

### Ukázka `templates/404.html`

```html
<!DOCTYPE html>
<html lang="cs">
<head>
    <meta charset="UTF-8">
    <title>Stránka nenalezena</title>
</head>
<body>
    <h1>404 — Stránka nenalezena 😢</h1>
    <p>Požadovaná stránka neexistuje.</p>
    <a href="/">Zpět na úvodní stránku</a>
</body>
</html>
```

> **Tip:** Chybové stránky jsou skvělá příležitost dát webu osobitý styl — místo generické chyby prohlížeče uvidí uživatel hezky naformátovanou stránku s navigací zpět.

# 1️⃣4️⃣ Dědičnost šablon (`{% extends %}`, `{% block %}`)

Většina stránek na webu sdílí **stejný layout** — navigaci, záhlaví, zápatí, CSS. Místo kopírování stejného kódu do každého HTML souboru vytvoříme **základní šablonu** (`base.html`) a ostatní šablony z ní **dědí**.

## Jak to funguje?

```
base.html (rodičovská šablona)
├── navbar, head, CSS, footer — společné pro všechny stránky
├── {% block content %}{% endblock %} — "prázdné místo" pro obsah
│
├── stranka1.html  →  {% extends "base.html" %}  +  {% block content %} vlastní obsah {% endblock %}
├── stranka2.html  →  {% extends "base.html" %}  +  {% block content %} jiný obsah {% endblock %}
└── ...
```

## Syntaxe

### Rodičovská šablona (`base.html`)
```html
<!DOCTYPE html>
<html lang="cs">
<head>
    <meta charset="UTF-8">
    <title>{% block title %}Můj web{% endblock %}</title>
    <link rel="stylesheet" href="{{ url_for('static', filename='styles/styly.css') }}">
</head>
<body>
    <nav>
        <a href="/home">Domů</a>
        <a href="/tabulka">Tabulka</a>
        <a href="/logout">Odhlásit</a>
    </nav>

    <div class="container">
        {% block content %}{% endblock %}
    </div>

    <footer>
        <p>&copy; 2026 SPŠ Kladno</p>
    </footer>
</body>
</html>
```

### Dceřiná šablona (`stranka.html`)
```html
{% extends "base.html" %}

{% block title %}Moje stránka{% endblock %}

{% block content %}
    <h1>Vítejte na mé stránce!</h1>
    <p>Toto je obsah specifický pro tuto stránku.</p>
{% endblock %}
```

### Klíčové prvky

| Příkaz | Kde | Význam |
|---|---|---|
| `{% block nazev %}...{% endblock %}` | `base.html` | Definuje **přepisovatelnou oblast** |
| `{% extends "base.html" %}` | dceřiná šablona | Zdědí vše z rodičovské šablony |
| `{% block nazev %}...{% endblock %}` | dceřiná šablona | **Přepíše** obsah bloku |
| `{{ super() }}` | dceřiná šablona | Zachová obsah rodiče + přidá vlastní |

### Kód

````python
# Route s dědičností šablon — Python kód se nemění!
from flask import Flask, render_template

app = Flask(__name__)

# Stačí render_template — dědičnost řeší Jinja2 v HTML souborech
@app.route('/profil')
def profil():
    return render_template("profil.html", jmeno="Adam", vek=18)

# profil.html dědí z base.html → automaticky má navbar, CSS, footer
````

### Ukázka `templates/profil.html`

```html
{% extends "base.html" %}

{% block title %}Profil uživatele{% endblock %}

{% block content %}
    <h1>Profil: {{ jmeno }}</h1>
    <p>Věk: {{ vek }}</p>
{% endblock %}
```

Výsledek: Prohlížeč obdrží kompletní HTML s navigací, CSS a zápatím z `base.html`, ale v sekci `content` bude profil uživatele.

> **Tip:** Dědičnost šablon je klíčový koncept pro reálné projekty. Změníte-li navigaci v `base.html`, promítne se to **automaticky na všech stránkách** — nemusíte měnit 20 souborů.

# 1️⃣5️⃣ Flash messages — jednorázové zprávy

Flash messages jsou **jednorázové zprávy**, které se uživateli zobrazí **jednou** po akci (např. „Úspěšně přihlášen!") a po obnovení stránky zmizí. Ideální pro potvrzení akcí, upozornění na chyby apod.

## Syntaxe

```python
from flask import flash, redirect, url_for

@app.route('/akce')
def akce():
    flash("Akce proběhla úspěšně!", "success")   # zpráva + kategorie
    flash("Pozor, něco se nepovedlo.", "error")
    return redirect(url_for('home'))
```

```html
<!-- V šabloně (nejlépe v base.html) -->
{% with messages = get_flashed_messages(with_categories=true) %}
    {% if messages %}
        {% for category, message in messages %}
            <div class="flash {{ category }}">{{ message }}</div>
        {% endfor %}
    {% endif %}
{% endwith %}
```

### Klíčové prvky

| Funkce | Kde | Popis |
|---|---|---|
| `flash("zpráva", "kategorie")` | Python | Uloží zprávu do session (zobrazí se na další stránce) |
| `get_flashed_messages()` | HTML (Jinja2) | Přečte a **smaže** všechny uložené zprávy |
| `with_categories=true` | HTML | Vrátí i kategorii (success, error, info…) |

> **⚠️ Pozor:** Flash messages vyžadují nastavený `app.secret_key` — stejně jako session, protože zprávy se ukládají právě do session.

### Kód

````python
# Flash messages — ukázka použití
from flask import Flask, render_template, request, redirect, url_for, flash

app = Flask(__name__)
app.secret_key = 'Muj_tajny_klic'

@app.route('/kontakt', methods=['GET', 'POST'])
def kontakt():
    if request.method == 'POST':
        jmeno = request.form.get('jmeno')
        zprava = request.form.get('zprava')

        if not jmeno or not zprava:
            flash("Vyplňte prosím všechna pole!", "error")
        else:
            # zde by se zpráva uložila do DB...
            flash(f"Děkujeme, {jmeno}! Vaše zpráva byla odeslána.", "success")
            return redirect(url_for('kontakt'))

    return render_template("kontakt.html")
````

### Ukázka `templates/kontakt.html` (s flash zprávami)

```html
{% extends "base.html" %}

{% block content %}
    <h1>Kontaktní formulář</h1>

    {# Zobrazení flash messages #}
    {% with messages = get_flashed_messages(with_categories=true) %}
        {% if messages %}
            {% for category, message in messages %}
                <div class="flash {{ category }}"
                     style="padding: 10px; margin: 10px 0;
                            background: {{ 'lightgreen' if category == 'success' else 'lightcoral' }};
                            border-radius: 5px;">
                    {{ message }}
                </div>
            {% endfor %}
        {% endif %}
    {% endwith %}

    <form method="post">
        <label>Jméno:</label><br>
        <input type="text" name="jmeno"><br><br>
        <label>Zpráva:</label><br>
        <textarea name="zprava"></textarea><br><br>
        <button type="submit">Odeslat</button>
    </form>
{% endblock %}
```

> **Tip:** Flash zprávy se ideálně umísťují do `base.html` — zobrazí se pak na jakékoli stránce, kam uživatele přesměrujeme. Kategorie (`success`, `error`, `info`) lze stylovat různými barvami přes CSS.

# 1️⃣6️⃣ JSON API (`jsonify`)

Moderní webové aplikace často komunikují pomocí **JSON** (JavaScript Object Notation) — posílají a přijímají data bez HTML. Takové route tvoří **API** (Application Programming Interface).

## Kdy se to hodí?

| Scénář | Příklad |
|---|---|
| **Mobilní aplikace** | Mobil pošle požadavek, server vrátí JSON s daty |
| **JavaScript (AJAX/fetch)** | Stránka se aktualizuje bez přenačtení |
| **Propojení systémů** | Dva servery si vyměňují data |
| **IoT / senzory** | Zařízení posílají data na server |

## Syntaxe

```python
from flask import jsonify, request

# Vrácení JSON odpovědi
@app.route('/api/data')
def api_data():
    return jsonify({"jmeno": "Adam", "vek": 18, "skola": "SPŠ Kladno"})

# Přijetí JSON dat z požadavku (POST)
@app.route('/api/prijem', methods=['POST'])
def api_prijem():
    data = request.get_json()  # parsuje JSON z těla požadavku
    jmeno = data.get('jmeno')
    return jsonify({"status": "ok", "prijato": jmeno})
```

| Funkce | Popis |
|---|---|
| `jsonify(dict)` | Převede Python slovník na JSON odpověď s hlavičkou `Content-Type: application/json` |
| `request.get_json()` | Přečte JSON data z těla POST požadavku |
| `request.args.get('klíč')` | Přečte parametry z URL (query string: `?klíč=hodnota`) |

### Kód

````python
# JSON API — příklady
from flask import Flask, jsonify, request

app = Flask(__name__)

# GET — vrátí seznam studentů jako JSON
@app.route('/api/studenti')
def api_studenti():
    studenti = [
        {"id": 1, "jmeno": "Adam", "rocnik": 4},
        {"id": 2, "jmeno": "Eva", "rocnik": 3},
        {"id": 3, "jmeno": "Petr", "rocnik": 4},
    ]
    return jsonify(studenti)
# → [{"id":1,"jmeno":"Adam","rocnik":4}, ...]

# GET s parametrem — filtrování podle ročníku
@app.route('/api/studenti/rocnik')
def api_studenti_rocnik():
    rocnik = request.args.get('r', type=int)  # ?r=4
    studenti = [
        {"id": 1, "jmeno": "Adam", "rocnik": 4},
        {"id": 2, "jmeno": "Eva", "rocnik": 3},
    ]
    if rocnik:
        studenti = [s for s in studenti if s["rocnik"] == rocnik]
    return jsonify(studenti)
# Příklad: /api/studenti/rocnik?r=4  → vrátí jen 4. ročník

# POST — přijme JSON data a odpoví
@app.route('/api/pridej', methods=['POST'])
def api_pridej():
    data = request.get_json()  # přečte JSON z těla požadavku
    if not data or 'jmeno' not in data:
        return jsonify({"error": "Chybí jméno"}), 400  # status 400
    return jsonify({"status": "ok", "zprava": f"Přidán: {data['jmeno']}"}), 201
````

# 📋 Shrnutí

| Koncept | Co dělá | Příkaz / Funkce |
|---|---|---|
| Vytvoření aplikace | Inicializace Flasku | `app = Flask(__name__)` |
| Route | Mapuje URL na funkci | `@app.route('/cesta')` |
| Šablona | Načte HTML soubor | `render_template('soubor.html')` |
| Statické soubory | Cesta k CSS/obrázkům | `url_for('static', filename='...')` |
| Proměnné v HTML | Předání dat do šablony | `render_template('s.html', x=hodnota)` |
| Formuláře | Čtení dat z POST | `request.form.get('pole')` |
| Upload souborů | Nahrání souboru | `request.files.get('file')` |
| URL path parametry | Dynamické cesty | `@app.route('/<int:id>')` |
| Query string parametry | Čtení `?klíč=hodnota` z URL | `request.args.get('klíč')` |
| Přesměrování | Redirect na jinou URL | `redirect(url_for('funkce'))` |
| Session | Udržení stavu přihlášení | `session['klíč'] = hodnota` |
| Hashování hesel | Bezpečné ukládání hesel | `bcrypt.hashpw(...)` |
| Grafy | Interaktivní vizualizace | `plotly.graph_objects` + `\|safe` |
| Chybové stránky | Vlastní 404/500 stránky | `@app.errorhandler(404)` |
| Dědičnost šablon | Společný layout pro stránky | `{% extends "base.html" %}` |
| Flash messages | Jednorázové zprávy uživateli | `flash("zpráva", "kategorie")` |
| JSON API | Vrácení dat ve formátu JSON | `jsonify({"klíč": "hodnota"})` |

### Jinja2 — šablonovací jazyk

| Zápis | Význam | Příklad |
|---|---|---|
| `{{ výraz }}` | Vypsání hodnoty | `{{ message }}` |
| `{% příkaz %}` | Řídící struktura | `{% if podmínka %}` |
| `{# komentář #}` | Komentář (nezobrazí se) | `{# TODO #}` |
| `\|safe` | Nevyescapuje HTML | `{{ graph_html\|safe }}` |
| `{% extends %}` | Dědičnost z rodičovské šablony | `{% extends "base.html" %}` |
| `{% block %}` | Přepisovatelná oblast | `{% block content %}...{% endblock %}` |
| `{% with %}` | Lokální proměnná | `{% with msgs = get_flashed_messages() %}` |

### Objekty `request`

| Atribut | Účel | Příklad |
|---|---|---|
| `request.method` | Zjistí HTTP metodu | `if request.method == 'POST'` |
| `request.form.get('x')` | Data z formuláře (POST) | Textová pole, selecty |
| `request.args.get('x')` | Query string (GET) | `?jmeno=Adam&vek=18` |
| `request.files.get('x')` | Nahraný soubor | Upload přes formulář |
| `request.get_json()` | JSON z těla požadavku | API POST data |

### HTTP metody

| Metoda | Účel | Příklad |
|---|---|---|
| `GET` | Získání dat (zobrazení stránky) | Otevření URL v prohlížeči |
| `POST` | Odeslání dat (formulář, JSON) | Submit formuláře |
| `PUT` | Aktualizace existujícího záznamu | API update |
| `DELETE` | Smazání záznamu | API delete |
